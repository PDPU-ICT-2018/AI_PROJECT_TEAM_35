# Importing Library

```python
#Importing Alpha_Vantage Library For Downloading Stock Data
from alpha_vantage.timeseries import TimeSeries
#Importing Pandas To perform ETL Operation On Stock Data
import pandas as pd
```

# Retrieving stock data using API 

```python
#Using Alpha_Vantage API's to call stock data
ts = TimeSeries(key = 'UGXITKR3V1V5SCNT', output_format = 'pandas')
#Downloading Stock Data of target company i.e. GoldmnanSachs and its compeititors
GS_data, GS_meta = ts.get_daily_adjusted(symbol = 'GS', outputsize = 'full')
MS_data, MS_meta = ts.get_daily_adjusted(symbol = 'MS', outputsize = 'full')
JPM_data, JPM_meta = ts.get_daily_adjusted(symbol = 'JPM', outputsize = 'full')
WFC_data, WFC_meta = ts.get_daily_adjusted(symbol = 'WFC', outputsize = 'full')
C_data, C_meta = ts.get_daily_adjusted(symbol = 'C', outputsize = 'full')
```

```python
BAC_data, BAC_meta = ts.get_daily_adjusted(symbol = 'BAC', outputsize = 'full')
```

```python
#Renaming data colunms to improve clarity of Data to layman
GS_data.rename(columns={'1. open': 'GS_open', '2. high': 'GS_high', '3. low': 'GS_low', '4. close': 'GS_close', '5. adjusted close':'GS_aclose', '6. volume': 'GS_volume'},inplace = True)
MS_data.rename(columns={'1. open': 'MS_open', '2. high': 'MS_high', '3. low': 'MS_low', '4. close': 'MS_close', '5. adjusted close':'MS_aclose', '6. volume': 'MS_volume'},inplace = True)
JPM_data.rename(columns={'1. open': 'JPM_open', '2. high': 'JPM_high', '3. low': 'JPM_low', '4. close': 'JPM_close', '5. adjusted close':'JPM_aclose', '6. volume': 'JPM_volume'},inplace = True)
WFC_data.rename(columns={'1. open': 'WFC_open', '2. high': 'WFC_high', '3. low': 'WFC_low', '4. close': 'WFC_close', '5. adjusted close':'WFC_aclose', '6. volume': 'WFC_volume'},inplace = True)
C_data.rename(columns={'1. open': 'C_open', '2. high': 'C_high', '3. low': 'C_low', '4. close': 'C_close', '5. adjusted close':'C_aclose', '6. volume': 'C_volume'},inplace = True)
BAC_data.rename(columns={'1. open': 'BAC_open', '2. high': 'BAC_high', '3. low': 'BAC_low', '4. close': 'BAC_close', '5. adjusted close':'BAC_aclose', '6. volume': 'BAC_volume'},inplace = True)
```

```python
#Removing useless colunms as they don't affect the price of Goldmam Sachs Stocks
GS_data = GS_data.drop(columns = ['7. dividend amount', '8. split coefficient'])
MS_data = MS_data.drop(columns = ['7. dividend amount', '8. split coefficient'])
JPM_data = JPM_data.drop(columns = ['7. dividend amount', '8. split coefficient'])
WFC_data = WFC_data.drop(columns = ['7. dividend amount', '8. split coefficient'])
C_data = C_data.drop(columns = ['7. dividend amount', '8. split coefficient'])
BAC_data = BAC_data.drop(columns = ['7. dividend amount', '8. split coefficient'])
```

# Concating all data

```python
#Concating all the data to create a common datasheet
trial_data = pd.concat([GS_data,MS_data,JPM_data,WFC_data,C_data,BAC_data], axis=1) 
```

```python
trial_data.head()
```

# Sorting data and saving in CSV file

```python
#Sorting The GoldmanSachs StockData in Ascending_Order
trial_data = trial_data.sort_index(axis = 0, ascending = True)
```

```python
trial_data.head()
```

```python
#Deleting first 33 rows as the, 
#Data points for each every data attributes like technical indicators exist from this dates.
trial_data = trial_data.iloc[33:]
```

```python
trial_data.to_csv('initial_data3.csv')
```
