# Importing Libraries

```python
#Importing Alpha_Vantage Library To Download Technical_Indicator 
from alpha_vantage.techindicators import TechIndicators
#Importing Pandas To perform ETL Operation On Indicator Data
import pandas as pd
```

# Reading CSV using pandas

```python
#Reading The StockData Created in Previous Step i.e. STEP-1
df = pd.read_csv('initial_data3.csv')
```

# Retrieving technical indicators from alphavantage API

```python
#Using Alpha_Vantage API's to call technical indicators on GoldmanSachs Data
ti = TechIndicators(key='UGXITKR3V1V5SCNT', output_format='pandas')
#Downloading Different Technical Indicator Data For GoldmanSachs Stock
GS_SMA7, GS_SMA7_META = ti.get_sma(symbol='GS', interval='daily', time_period=7, series_type = 'close')
GS_SMA21, GS_SMA21_META = ti.get_sma(symbol='GS', interval='daily', time_period=21, series_type = 'close')
GS_EMA21, GS_EMA21_META = ti.get_ema(symbol='GS', interval='daily', time_period=21, series_type = 'close')
GS_MACD, GS_MACD_META = ti.get_macd(symbol='GS', interval='daily', series_type = 'close')
GS_BBANDS, GS_BBANDS_META = ti.get_bbands(symbol='GS', interval='daily', time_period=20,  series_type = 'close')
```

```python
GS_STOCH, GS_STOCH_META = ti.get_stoch(symbol='GS', interval='daily')
GS_RSI, GS_RSI_META = ti.get_rsi(symbol='GS', interval='daily',series_type = 'close',time_period = 14)
GS_ADX, GS_ADX_META = ti.get_adx(symbol='GS', interval='daily',time_period = 14)
GS_CCI, GS_CCI_META = ti.get_cci(symbol='GS', interval='daily',time_period = 20)
GS_AROON, GS_AROON_META = ti.get_aroon(symbol='GS', interval='daily',time_period = 25)
```

```python
GS_AD, GS_AD_META = ti.get_ad(symbol='GS', interval='daily')
GS_OBV, GS_OBV_META = ti.get_obv(symbol='GS', interval='daily')
```

# Renaming and sorting colums

```python
#Renaming column name
GS_SMA7 = GS_SMA7.rename(columns = {'SMA':'GS_SMA7'})
GS_SMA21 = GS_SMA21.rename(columns = {'SMA':'GS_SMA21'})
GS_EMA21 = GS_EMA21.rename(columns = {'EMA':'GS_EMA21'})
GS_MACD = GS_MACD.rename(columns = {'MACD':'GS_MACD', 'MACD_Signal':'GS_MACD_sig','MACD_Hist':'GS_MACD_Hist'})
GS_BBANDS = GS_BBANDS.rename(columns = {'Real Upper Band':'GS_Upper', 'Real Lower Band':'GS_Lower','Real Middle Band':'GS_Middle'})
GS_STOCH = GS_STOCH.rename(columns = {'SlowK':'GS_SlowK','SlowD':'GS_SlowD'})
GS_RSI = GS_RSI.rename(columns = {'RSI':'GS_RSI'})
GS_ADX = GS_ADX.rename(columns = {'ADX':'GS_ADX'})
GS_CCI = GS_CCI.rename(columns = {'CCI':'GS_CCI'})
GS_AROON = GS_AROON.rename(columns = {'Aroon Up':'GS_aroon_up','Aroon Down':'GS_aroon_down'})
GS_AD = GS_AD.rename(columns = {'Chaikin A/D':'GS_AD'})
GS_OBV = GS_OBV.rename(columns = {'OBV':'GS_OBV'})
```

```python
#Sorting Technical Data For GoldmanSachs Data in Ascending_Order
GS_STOCH = GS_STOCH.sort_index(axis = 0, ascending = True)
GS_AROON = GS_AROON.sort_index(axis = 0, ascending = True)
GS_MACD = GS_MACD.sort_index(axis = 0, ascending = True)
GS_BBANDS = GS_BBANDS.sort_index(axis = 0, ascending = True)
```

```python
#Removing certain number of columns to achieve date where every data point exists.
Ti_1 = Ti_1.iloc[27:]
Ti_2 = Ti_2.iloc[33:]
```

```python
#Concatenating TechnicalIndicator Data 
Ti_Final = pd.concat([Ti_1,Ti_2], axis = 1)
```

```python
#Resetting Index For Final TechnicalIndicator Data 
Ti_Final = Ti_Final.reset_index()
Ti_Final = Ti_Final.drop(columns = 'date')
```

```python
#Concatenating StockData and TechnicalIndicator Data To Create Final DataSet
Final_Data1 = pd.concat([df,Ti_Final], axis = 1)
#Copying data into new dataframe for clarity purposes
Final_Data2 = Final_Data1
```
